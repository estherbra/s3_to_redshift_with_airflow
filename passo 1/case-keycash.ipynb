```python
import boto3
import json
import pandas as pd
import os
import glob
from botocore.exceptions import NoCredentialsError
```

```python
#importando chaves de acesso para realizar conexão com a AWS
ACCESS_KEY = '*'
SECRET_KEY = '*'
```

```python
def upload_to_aws(local_file, bucket, s3_file):
    s3 = boto3.client('s3', aws_access_key_id=ACCESS_KEY,
                      aws_secret_access_key=SECRET_KEY)

    try:
        s3.upload_file(local_file, bucket, s3_file)
        print("Upload Successful")
        return True
    except FileNotFoundError:
        print("The file was not found")
        return False
    except NoCredentialsError:
        print("Credentials not available")
        return False
```

```python
#convertendo json para csv e salvando em novo arquivo
 
#utilizando modulo glob para retornar arquivos que correspondam ao padrão especificado
json_files = glob.glob('./json_files/*.json')

for file_path in json_files:
    file_name = os.path.basename(file_path)
    csv_name = file_name.replace(".json", ".csv")
    json_file = open(file_path)
    clients = json.load(json_file)["clientes"]
    f = pd.json_normalize(clients)
    csv_file = f.to_csv("./csv_files/" + csv_name , index = False, header=False)
```

```python
#criando for loop para fazer upload de cada um dos 50 csv files e para criação de respectivos objetos no bucket:
csv_files = glob.glob('./csv_files/*.csv')

for file_path in csv_files:
  file_name = os.path.basename(file_path)
  upload_to_aws(file_path,'bucket-case-keycash', file_name)
```
